```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/content/simulated_paper_reviews.xlsx")
df
```

```python
# Grain 1: What is the average review score?
average_score = np.round(df["Review Score"].mean(), 2)
print("Average Review Score:", average_score)
```

```python
# Grain 2: How many papers were accepted?
accepted_count = (df["Final Decision"] == "Accept").sum()
print("Number of Accepted Papers:", accepted_count)
```

```python
# Grain 3: How many papers were rejected?
rejected_count = (df["Final Decision"] == "Reject").sum()
print("Number of Rejected Papers:", rejected_count)
```

```python
# Grain 4: What is the most common recommendation?
most_common_recommendation = df["Recommendation"].mode()[0]
print("Most Common Recommendation:", most_common_recommendation)
```

```python
# Grain 5: What is the average score for accepted papers?
accepted_df = df[df["Final Decision"] == "Accept"]
average_score_accepted = np.round(accepted_df["Review Score"].mean(), 2)
print("Average Score for Accepted Papers:", average_score_accepted)
```

```python
# Grain 6: What is the average score for rejected papers?
rejected_df = df[df["Final Decision"] == "Reject"]
average_score_rejected = np.round(rejected_df["Review Score"].mean(), 2)
print("Average Score for Rejected Papers:", average_score_rejected)
```

```python
# Grain 7: Which paper had the highest review score?
max_score_row = df.loc[df["Review Score"].idxmax()]
print("Paper with Highest Score:", max_score_row["Paper ID"], "-", max_score_row["Title"])
```

```python
# Grain 8: Which paper had the lowest review score?
min_score_row = df.loc[df["Review Score"].idxmin()]
print("Paper with Lowest Score:", min_score_row["Paper ID"], "-", min_score_row["Title"])
```

```python
# Grain 9: How many unique reviewers participated?
unique_reviewers = df["Reviewer Name"].nunique()
print("Unique Reviewers:", unique_reviewers)
```

```python
# Grain 10: What is the most common final decision?
most_common_decision = df["Final Decision"].mode()[0]
print("Most Common Final Decision:", most_common_decision)
```

```python
# Grain 11: What is the range of review scores?
min_score = df["Review Score"].min()
max_score = df["Review Score"].max()
print(f"Review Score Range: {min_score} to {max_score}")
```

```python
# Grain 12: What is the median review score?
median_score = df["Review Score"].median()
print("Median Review Score:", median_score)
```

```python
# Grain 13: What is the standard deviation of review scores?
std_dev_score = np.round(df["Review Score"].std(), 2)
print("Standard Deviation of Review Scores:", std_dev_score)
```

```python
# Grain 14: What is the count of each recommendation?
recommendation_counts = df["Recommendation"].value_counts()
print("Recommendation Counts:")
print(recommendation_counts)
```

```python
# Grain 15: What is the count of each final decision?
decision_counts = df["Final Decision"].value_counts()
print("Final Decision Counts:")
print(decision_counts)
```

```python
# Grain 16: What is the average time from submission to review?
avg_days_to_review = (df["Review Date"] - df["Submission Date"]).mean().days
print("Average Days from Submission to Review:", avg_days_to_review)
```

```python
# Grain 17: What is the average time from review to decision?
avg_days_to_decision = (df["Decision Date"] - df["Review Date"]).mean().days
print("Average Days from Review to Decision:", avg_days_to_decision)
```

```python
# Grain 18: Which paper had the longest review comment?
longest_comment_row = df.loc[df["Comments"].str.len().idxmax()]
print("Paper with Longest Comment:", longest_comment_row["Paper ID"])
```

```python
# Grain 19: Who reviewed the most papers?
most_reviews = df["Reviewer Name"].value_counts().idxmax()
most_reviews_count = df["Reviewer Name"].value_counts().max()
print(f"Most Active Reviewer: {most_reviews} ({most_reviews_count} papers)")
```

```python
# Grain 20: Which paper had a score closest to the average?
average_score = df["Review Score"].mean()
closest_to_average = df.iloc[(df["Review Score"] - average_score).abs().idxmin()]
print("Paper Closest to Average Score:", closest_to_average["Paper ID"], "-", closest_to_average["Title"])
```
